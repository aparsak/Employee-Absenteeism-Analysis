```python
# convert it to a date-time
df["Date"] = pd.to_datetime(df["Date"], format = '%d/%m/%Y')
# display the first date
df["Date"][0]
```

```python
import pandas as pd

# Assuming df is your DataFrame

# Convert "Date" column to datetime
df["Date"] = pd.to_datetime(df["Date"], format='%d/%m/%Y')

# Extract day, week, month, a
df["Month"] = df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df["Weekday"] = df["Date"].dt.weekday  # Add this line to extract the weekday

# Function to extract season
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Autumn'
    else:
        return 'Winter'

# Apply function to get season
df["Season"] = df["Month"].apply(get_season)

# Drop the original "Date" column
df.drop(columns=["Date"], inplace=True)

# Display the DataFrame
df.head()
```

<b> 2. تحلیل مصرف کنندگان الکل و سیگار </b>

با تجزیه و تحلیل تأثیر مصرف الکل یا دخانیات بر غیبت کارکنان شروع کنیم. با توجه به اینکه سیگار کشیدن و مصرف مکرر الکل تأثیر منفی بر شرایط سلامتی دارند، انتظار داریم که برخی از بیماری‌ها در سیگاری‌ها و مصرف‌کنندگان الکل بیشتر از دیگران باشند.

```python
# Convert "Social_Drinker_Status" and "Social_Smoker_Status" to numeric values
df["Social_Drinker_Status"] = df["Social_Drinker_Status"].map({"No": 0, "Yes": 1})
df["Social_Smoker_Status"] = df["Social_Smoker_Status"].map({"No": 0, "Yes": 1})

# Calculate the percentage of employees who are social drinkers
social_drinkers_percentage = (df["Social_Drinker_Status"].sum() / len(df)) * 100

# Calculate the percentage of employees who are social smokers
social_smokers_percentage = (df["Social_Smoker_Status"].sum() / len(df)) * 100

print(f"Percentage of employees who are social drinkers: {social_drinkers_percentage:.2f}%")
print(f"Percentage of employees who are social smokers: {social_smokers_percentage:.2f}%")
```

```python
from scipy.stats import yeojohnson

df['Absent_Hours'], _ = yeojohnson(df['Absence_Duration_Hours'] + 1)


sns.jointplot(x='Distance_to_Work', y='Absent_Hours', data=df, kind='reg')
plt.xlabel('Distance to Work')
plt.ylabel('Absent Hours')
plt.show()

sns.jointplot(x='Transportation_Expense', y='Absent_Hours', data=df, kind='reg')
plt.xlabel('Transportation Expense')
plt.ylabel('Absent Hours')
plt.show()
```
